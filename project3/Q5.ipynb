```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
accepted = []
p = [0.06]*5+[0.15, 0.13, 0.14, 0.15, 0.13]+[0]*10
q = [0.05]*20
c = 3
N = 10000
```

```python
for i in range(N):
    j = np.random.randint(20)
    Random_model = np.random.uniform(0, 1)
    if Random_model < p[j]/(c*q[j]):
        accepted.append(j+1)
```

```python
plt.figure()
plt.hist(accepted, bins=range(21), rwidth=0.8, edgecolor='black', color='b',align='left', label='Accept-Reject Sample')
plt.plot(range(1, 21), [x*N/c for x in p], marker='o', color='r', label='distribution pj')
plt.xlabel('Random Variable')
plt.ylabel('Frequency')
plt.title('Distribution of p')
plt.legend()
plt.show()
```

```python
print('E[x] =', np.mean(accepted))
print('Var[x] =', np.var(accepted, ddof=1))
print('Efficiency =', len(accepted)/N)
```
