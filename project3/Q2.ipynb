```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
lamda = 120
N = 10000
p = lamda/N
```

```python
poisson = []
prob = (lamda ** 79)*np.exp(-lamda)/np.math.factorial(79)
for i in range(16, 32):
    number = 0
    for j in range(5):
        x = 5*i+j
        prob *= (lamda/x)
        number += 1000 * prob
    poisson.append(int(number))
```

```python
bernoulli_result = []
for i in range(1000):
    count = 0
    for j in range(N):
        Random_model = np.random.uniform(0, 1)
        if Random_model <= p:
            count += 1
    bernoulli_result.append(count)
```

```python
plt.figure(1)
plt.hist(bernoulli_result, bins=range(min(bernoulli_result) - 1, max(bernoulli_result) + 2, 5), rwidth=0.8, alpha=0.4, color='b', label='bernoulli trials',  align='left')
plt.plot(range(80, 160, 5), poisson, marker='o', color='r', label='theoretical')
plt.xlabel('Number of Cars')
plt.ylabel('Frequency')
plt.title('Bernoulli trails')
plt.legend()
```

```python
inv_transform_result = []
for times in range(1000):
    U = np.random.uniform(0, 1)
    i = 0
    P = np.exp(-lamda)
    F = P
    while 1:
        if U < F:
            inv_transform_result.append(i)
            break
        else:
            P = (lamda*P)/(i+1)
            F += P
            i += 1
```

```python
plt.figure(2)
plt.hist(inv_transform_result, bins=range(min(inv_transform_result) - 1, max(inv_transform_result) + 2, 5), rwidth=0.8, alpha=0.4, color='b', label='inverse transform',  align='left')
plt.plot(range(80, 160, 5), poisson, marker='o', color='r', label='theoretical')
plt.xlabel('Number of Cars')
plt.ylabel('Frequency')
plt.title('Inverse transform')
plt.legend()
plt.show()
```
