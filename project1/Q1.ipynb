```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N_tosstime = 50
N_prob = 0.5
N_repeat = 1000
N_head = 0
N_tail = 0
N_longest = 0
record = np.empty([N_tosstime,1])
```

```python
for i in range(0,N_tosstime):
    if np.random.random() < N_prob:
        N_head += 1
        N_tail += 1
        record[i] = 1
        if N_tail >= N_longest:
            N_longest = N_tail
            
    else:
        N_tail = 0
        record[i] = 0
```

```python
print("Number of heads is:" ,str(N_head))
print("Longest run of heads is:" ,str(N_longest))
```

```python
plt.hist(record, 2, facecolor="blue", edgecolor="black", alpha=0.75)
plt.ylim([15, 35])
plt.xlabel("tail = 0; head = 1")
plt.ylabel("Number")
plt.title("Bernoulli result")
```

```python
total_head = []
for i in range(0,N_repeat):
    N_head = 0
    for j in range(0,N_tosstime):
        if np.random.random()< N_prob:
            N_head +=1
    total_head.append(N_head)
```

```python
x = np.arange(10,40)
plt.hist(total_head, x, facecolor="blue", edgecolor="black", alpha=0.75)
plt.xlabel("Each showing number of heads in 50 flips")
plt.ylabel("Number of heads")
plt.title("1000 times")
```
