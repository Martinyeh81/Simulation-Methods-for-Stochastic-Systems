```python
import numpy as np
from matplotlib import pyplot as plt
from scipy.stats import norm
```

```python
x = np.linspace(-5, 5, 100)
f = 0.4 * np.array(norm(-1, 1).pdf(x)) + 0.6 * np.array(norm(1, 1).pdf(x))
p = 0.6
result = []
for i in range(10000):
    if np.random.rand() < p:
        result.append(np.random.normal(1,1))
    else:
        result.append(np.random.normal(-1,1))
plt.figure()
plt.hist(result, bins=50, density=1, label='Simulation', color='b')
plt.plot(x, f.tolist(), label='Theoretical', color='r')
plt.legend()
plt.show()
```
