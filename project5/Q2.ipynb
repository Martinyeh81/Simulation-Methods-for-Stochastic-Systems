```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def main(x,y):
    r = [x, y]
    result_buffer1 = []
    result_buffer2 = []
    result_output1 = []
    result_output2 = []
    for p in np.arange(0.01,1,0.01):
        buffer1 = 0
        buffer2 = 0
        output1 = 0
        output2 = 0
        for times in range(1000):
            if buffer1 and buffer2:
                select1 = np.random.uniform()
                select2 = np.random.uniform()

                if select1 < r[0] and select2 < r[0]:
                    output1 += 1
                    if np.random.uniform() < 0.5:
                        buffer1 -= 1
                    else:
                        buffer2 -= 1

                elif select1 >= r[0] and select2 >= r[0]:
                    output2 += 1
                    if np.random.uniform() < 0.5:
                        buffer1 -= 1
                    else:
                        buffer2 -= 1

                else:
                    output1 += 1
                    output2 += 1
                    buffer1 -= 1
                    buffer2 -= 1

            elif buffer1:
                buffer1 -= 1
                if np.random.uniform() < r[0]:
                    output1 += 1
                else:
                    output2 += 1

            elif buffer2:
                buffer2 -= 1
                if np.random.uniform() < r[0]:
                    output1 += 1
                else:
                    output2 += 1

            if np.random.uniform() < p:
                buffer1 += 1
            if np.random.uniform() < p:
                buffer2 += 1
        result_buffer1.append(buffer1)
        result_buffer2.append(buffer2)
        result_output1.append(output1/1000)
        result_output2.append(output2/1000)

    plt.figure()
    plt.plot(np.arange(0.01,1,0.01), result_output1)
    plt.xlabel('Probability '+'r=['+str(x)+', '+str(y)+']')
    plt.ylabel('Number of packet per minute')

    plt.figure()
    plt.plot(np.arange(0.01, 1, 0.01), result_output2)
    plt.xlabel('Probability '+'r=['+str(x)+', '+str(y)+']')
    plt.ylabel('Number of packet per minute')
    plt.show()
```

```python
if __name__ == '__main__':
    main(0.5,0.5)
    main(0.75,0.25)
```
